```python
from nltk.tokenize import TweetTokenizer

import matplotlib.pyplot as plt
##install wordcloud
from wordcloud import WordCloud, STOPWORDS
import string
import pandas as pd
import numpy as np
from nltk.corpus import wordnet as wn
```

```python
def tokenize(text):
    #tokenization
    tknzr = TweetTokenizer()
    tokens = tknzr.tokenize(text)
    tokens = list(dict.fromkeys([wn.morphy(i) for i in tokens]))
    tokens = [i for i in tokens if i!=None]
    return tokens

def removeStp(tokens):
    tokens = [str(i).lower() for i in tokens]
    tokens = [i for i in tokens if i not in STOPWORDS]
    tokens = [i for i in tokens if i not in string.punctuation]
    tokens = [i for i in tokens if 'https' not in i]
    tokens = [i for i in tokens if '…' not in i]
    tokens = [i for i in tokens if 'rt' not in i]
    tokens = [i for i in tokens if i !='’']
    tokens = [i for i in tokens if len(i)!=1]
    return tokens


def createDF(tokens_wo_stp):
    word_unique = list(dict.fromkeys(tokens_wo_stp))
    freq = []
    for word in word_unique:
        freq+=[len([i for i in tokens_wo_stp if i==word])]
    df = pd.DataFrame(columns = ['freq','word'])
    df['freq']=freq
    df['word']=word_unique
    return df
    
def showCloud(df):
    dictionary = df.set_index('word').to_dict()['freq']
    wordcloud = WordCloud().generate_from_frequencies(dictionary)

    plt.figure(figsize = (10, 10), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
  
    plt.show()
```

```python
text=''
with open('financial.txt','r') as f:
    for line in f:
        text+=line

tokens = tokenize(text)
tokens_wo_stp = removeStp(tokens)
df = createDF(tokens_wo_stp)
#df.to_csv('forWordCloud.csv',index=False)
showCloud(df)
```
